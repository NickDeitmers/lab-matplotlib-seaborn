#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
x = np.arange(0,100)
y1 = x ** 2

fig3, axes3 = plt.subplots(1, 2, figsize = (10, 5))
axes3[0].plot(x, y1)
axes3[0].set_title("Left graph", fontsize = 15)

axes3[1].plot(x, np.e**x)
axes3[1].set_title("Right graph", fontsize = 15)
axes3[1].set_yscale(value = 'log')
```

#### As a bonus challenge, try to add a legend to the plot.

```python
# your code here

x = np.arange(0,100)
y1 = x ** 2

fig3, axes3 = plt.subplots(1, 2, figsize = (10, 5))
axes3[0].plot(x, y1, label = "normal scale")
axes3[0].set_title("Left graph", fontsize = 15)

axes3[1].plot(x, np.e**x, label = "logaritmic scale")
axes3[1].set_title("Right graph", fontsize = 15)
axes3[1].set_yscale(value = 'log')

for i in axes3:
    i.legend()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
steps_month = fitbit.groupby(["Months_encoded"])["Steps"].median().reset_index()

# Create figure
fig4, axes4 = plt.subplots(2, 1, figsize = (12, 4))

# by month
axes4[0].bar(steps_month["Months_encoded"], steps_month["Steps"])
plt.sca(axes4[0])
plt.xticks(ticks=range(1,13), labels = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sept", "Oct", "Nov", "Dec"])

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
steps_weekend = fitbit.groupby(["Work_or_Weekend"])['Steps'].median().reset_index()

axes4[1].bar(steps_weekend["Work_or_Weekend"], steps_weekend["Steps"])
plt.sca(axes4[1])
plt.xticks(ticks=range(2), labels = ["Work", "Weekend"])

# Save it
fig4.savefig("figures.png") 
```
